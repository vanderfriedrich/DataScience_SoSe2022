```python
import numpy as np
import matplotlib.pyplot as plot
from scipy import stats
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
df = pd.read_csv("https://raw.githubusercontent.com/vanderfriedrich/DataScience_SoSe2022/main/koerper-schuhe.csv")

x = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
print(x)
print(y)
stats.describe(y)
plot.plot(y)
plot.show()

xTrain, xTest, yTrain, yTest = train_test_split(x,y,test_size=1/7, random_state=0)
print(xTrain)
print(yTrain)

linearRegressor = LinearRegression()
linearRegressor.fit(xTrain, yTrain)
print(linearRegressor)
yPrediction = linearRegressor.predict(xTest)
print(yPrediction)

plot.scatter(xTrain, yTrain, color='green')
plot.plot(xTrain, linearRegressor.predict(xTrain), color='violet')
plot.title('Körpergröße in Zusammenhang zur Schuhgröße')
plot.xlabel('Schuhgröße')
plot.ylabel('Körpergröße')

plot.show()
```
